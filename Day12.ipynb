```python
input12 = [list(line) for line in open('input12.txt').read().splitlines()]

S = (0,0)
E = (0,0)
a_list = []
adj = set()

for i, line in enumerate(input12):
    for j, node in enumerate(line):
        if node == 'S':
            S = (i,j)
            input12[i][j] = 'a'
        if node == 'E':
            E = (i,j)
            input12[i][j] = 'z'

for i, line in enumerate(input12):
    for j, node in enumerate(line):
        if node == 'a':
            a_list.append((i,j))
        if i!=len(input12)-1:
            adj.add(((i,j), (i+1,j), ord(input12[i+1][j])-ord(node)))
        if j!=len(line)-1:
            adj.add(((i,j), (i,j+1), ord(input12[i][j+1])-ord(node)))

def get_dist(start):
    adj_ = adj.copy()
    dist = {}
    remove_list = []
    
    for pair in adj_:
        if (pair[0]==start and pair[2]<=1) or (pair[1]==start and pair[2]>=-1):
            dist[pair[0] if pair[0]!=start else pair[1]] = 1
            remove_list.append(pair)
    while len(remove_list) > 0:
        adj_.remove(remove_list.pop())
    
    prev_len = -1
    while len(adj_) != prev_len:
        prev_len = len(adj_)
        clone = dist.copy()
        for node in clone:
            if node != E:
                for pair in adj_:
                    if (pair[0]==node and pair[2]<=1) or (pair[1]==node and pair[2]>=-1):
                        new = pair[0] if pair[0]!=node else pair[1]
                        if new in dist:
                            dist[new] = min(dist[node]+1, dist[new])
                        else:
                            dist[new] = dist[node]+1
                        remove_list.append(pair)
                del dist[node]
                while len(remove_list) > 0:
                    adj_.remove(remove_list.pop())
    if E in dist:
        return dist[E]
    else:
        return 9999  

print(get_dist(S))
print(min(map(get_dist,a_list)))
```
